# Punjab Consumption Analysis

This project analyzes district-level consumption patterns in Punjab using NSSO data, including preprocessing, visualization, and mapping.

```python
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
```

```python
data = pd.read_excel("Punjab_ds.xlsx")
data.head()
```

```python
print(data.isnull().sum())
data = data.dropna()
```

```python
data['total_consumption'] = (
    data['pulsestot_q'] +
    data['milkprott_q'] +
    data['potato_q'] +
    data['sugaros_q'] +
    data['eggsno_q']
)
```

```python
district_data = data.groupby('District')['total_consumption'].sum().reset_index()
district_data
```

```python
plt.figure(figsize=(10,5))
plt.bar(district_data['District'], district_data['total_consumption'])
plt.xticks(rotation=90)
plt.title("Total Consumption by District")
plt.show()
```

```python
gdf = gpd.read_file("indian_districts.shp")
merged = gdf.merge(district_data, on="District")
merged.plot(column='total_consumption', cmap='Reds', legend=True)
plt.title("Consumption by District")
plt.show()
```

## Key Insights

- Significant variation in consumption across districts
- Higher consumption in districts like Ludhiana and Amritsar
- Lower consumption in districts like Mansa and Faridkot
- Clear spatial patterns across Punjab
